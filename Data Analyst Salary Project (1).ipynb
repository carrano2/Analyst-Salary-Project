```python
%%sql

Select *
from data_jobs
```

```python
%%sql

select distinct
    job_title, experience_level, salary_in_usd
from 
    data_jobs
where
    job_title like '%Analyst%' and company_location = 'US'
```

I will now create a list of the average salary of Analyst roles in the US based off of job_title and experience level from the dataset.  Then I will graph it.

```python
%%sql

create table if not exists Analyst_Positions_List as
select
    job_title, avg(salary_in_usd) as average_salary, experience_level
from
    data_jobs
where
    job_title like '%Analyst%' and company_location = 'US'
group by 
    job_title, experience_level
order by 
    job_title, average_salary
```

```python
%%sql
select *
from Analyst_Positions_List
```

I cannot have a graph with two x axis variables in order to properly visualize this data in the above table, therefore I will have to concatenate job_title and experience_level

```python
%%sql
create table if not exists
    Bar_Graph_Analyst as
select
    job_title || '--' || experience_level as Job,
    average_salary
from
    Analyst_Positions_list
```

I will now query a table from SQL and convert it to Python as a dataframe in order to graph it.

```python
%%sql sqlres <<
select * 
from Bar_Graph_Analyst

```

```python
df = sqlres.DataFrame()
print(df)
```
